```python
# 1. DEFINICIÓN DE LA CLASE
class InstrumentosMusicales:
    def __init__(self, nombre, familia, material, origen):
        # Atributos de la clase
        self.nombre = nombre
        self.familia = familia
        self.material = material
        self.origen = origen

    # Función 1: afinar
    def afinar(self):
        print(f"\n[ACCIÓN: AFINAR] Ajustando el/la {self.nombre}...")
        print(f"Las partes de {self.material} están listas para sonar.")

    # Función 2: tocar
    def tocar(self):
        print(f"[ACCIÓN: TOCAR] Interpretando una melodía con el/la {self.nombre}.")
        print(f"¡Qué buen sonido tiene esta pieza de la familia {self.familia}!")

    # Función 3: guardar
    def guardar(self):
        print(f"[ACCIÓN: GUARDAR] El instrumento de origen {self.origen} ha sido puesto en su estuche.")

# 2. INTERACCIÓN PARA INTRODUCIR ATRIBUTOS
print("--- REGISTRO DE INSTRUMENTOS MUSICALES ---")

# Creación del primer objeto (instrumento1)
print("\nDatos para el Instrumento 1:")
n1 = input("Nombre: ")
f1 = input("Familia: ")
m1 = input("Material: ")
o1 = input("Origen: ")
instrumento1 = InstrumentosMusicales(n1, f1, m1, o1)

# Creación del segundo objeto (instrumento2)
print("\nDatos para el Instrumento 2:")
n2 = input("Nombre: ")
f2 = input("Familia: ")
m2 = input("Material: ")
o2 = input("Origen: ")
instrumento2 = InstrumentosMusicales(n2, f2, m2, o2)

# 3. EJECUCIÓN DE FUNCIONES
print("\n" + "="*40)
print("DEMOSTRACIÓN DE FUNCIONES DE LOS OBJETOS")
print("="*40)

# Interactuando con instrumento1
print(f"\n--- Probando: {instrumento1.nombre} ---")
instrumento1.afinar()
instrumento1.tocar()
instrumento1.guardar()

# Interactuando con instrumento2
print(f"\n--- Probando: {instrumento2.nombre} ---")
instrumento2.afinar()
instrumento2.tocar()
instrumento2.guardar()

print("Programa hecho por Caleb Dorado 0039, NL = 14")
```
